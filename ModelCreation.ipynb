```python
#Load raw data
import pandas as pd
import numpy as np
df = pd.read_csv('DataSets/EngineeredData.csv')
print(df.shape)
df.head()
```

```python
#Split the label feature from the other features
y = df['label']
X = df.drop(['label'], axis=1)
```

```python
#Split training and testing set
from sklearn.model_selection import train_test_split

X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, 
                                                    test_size = 0.20, random_state = 6)
```

```python
#Train the Decision Tree model
from sklearn.tree import DecisionTreeClassifier

# instantiate the model 
tree = DecisionTreeClassifier(max_depth = 3)
# fit the model 
tree.fit(X_train_val, y_train_val)
y_test_tree = tree.predict(X_test)
```

```python
#Report accuracy
from sklearn.metrics import accuracy_score

print("Decision Tree accuracy is {:.4f}".format(accuracy_score(y_test, y_test_tree)))
```

```python
#Save model
import pickle

with open('model.pkl', 'wb') as f:
    pickle.dump(tree, f)
```
